```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
# 주어진 데이터셋의 일부 카테고리 데이터만 추출하므로 카테고리 사전에 설정
cates = ["rec.motorcycles", "rec.sport.baseball", "comp.graphics", "comp.windows.x", "talk.politics.mideast", "soc.religion.christian", "sci.electronics", "sci.med"]
```

```python
# 설정해준 카테고리의 데이터들만 추출
# 해당 뉴스그룹에서 remove는 headers와 footers quotes는 제거하고 본문만 가져온다 그런 말? 
# categories는 cates라는 총 8개만 쓴다는 것. 위에 변수로 저장해놓음. 
# subset은 Train과 test나누지 않고 다 가져오기
# 왜냐면 우리는 lda를 쓰기 때문에. 
news_df = fetch_20newsgroups(subset="all", remove=("headers", "footers", "quotes"), categories = cates, random_state=12)
```

```python
# CountVectorizer로 텍스트 데이터를 단어 빈도수에 기반해 벡터화시키기(fit_trasform까지 같이하기)
count_vect = CountVectorizer(max_df = 0.95, max_features = 1000, min_df = 2, stop_words="english", ngram_range=(1, 2))
ftr_vect = count_vect.fit_transform(news_df.data)
```

```python
count_vect.vocabulary_
```

```python
ftr_vect
```

```python
#LDA클래스를 이용해서 피처 벡터화시킨것을 토픽모델링 시키기
# 8개의 주제만 뽑았으니 n_components(토픽개수) 8로 설정
lda = LatentDirichletAllocation(n_components=8, random_state=42)
lda.fit(ftr_vect)
```

```python
print(lda.components_.shape)
print(lda.components_)
```

# 위 결과의 설명
shape가 8개의 행, 1000개의 칼럼으로 이루어진 array를 반환함.  
> 행은 각각의 토픽  
열은 단어들을 Vectorize한 feature들

## fit 호출
: 토픽별 단어들의 분포 를 반환
## fit + transform 호출
: 문서별 토픽들의 분포 까지 반환

```python
# 요 함수에서 lda_model이란, 벡터화시킨 텍스트 데이터를 fit까지만 적용한 모델
def display_topic_words(lda_model, feature_names, num_top_words):
    for topic_idx, topic in enumerate(lda_model.components_):
        print("\nTopic #", topic_idx+1)
        # Topic별로 1000개의 단어들(features)중에서 높은 값 순으로 정렬 후 index를 반환해줌
        # argsort()는 default가 오름차순(1, 2, 3,..) 그래서[::,-1]로 내림차순으로 바꾸기
        topic_word_idx = topic.argsort()[::-1]
        top_idx = topic_word_idx[:num_top_words]
        
        # CountVectorizer 함수 할당시킨 객체에 get_feature_names()로 벡터화시킨 feature(단어들) 볼 수 있음.
        # 이 벡터화시킨 단어들(features)은 숫자-알파벳순으로 정렬되며, 단어들 순서는 fit_transform시키고 난 이후에도 동일! 
        # "문자열".join함수로 특정 문자열 사이에 끼고 문자열 합쳐줄 수 있음
        feature_concat = "+".join([str(feature_names[i])+"*"+str(round(topic[i], 1)) for i in top_idx])
        print(feature_concat)
feature_names = count_vect.get_feature_names_out()
display_topic_words(lda, feature_names, 15)
```

## 7862개의 행과 8개의 열

> 주어진 내장 텍스트데이터의 문서이름에는 카테고리가 labeling 되어 있음  
따라서 카테고리가 무엇인지 아는 상태이니까 어떤 문서들이 어떤 토픽들이 높은지 확인해볼 수 있음. 그렇기 때문에 그 토픽들이 각각 무엇을 내용으로 하는지 추측가능함. 실제로는 라벨링이 되있지 않기 때문에. 각각의 커뮤니티(우리의 프로젝트)에서의 가장 많이 나오는 단어들을 확인할 수 있고, 이 것이 긍정 부정을 확인할 수 있는 말뭉치를 돌려서 올리는 것으로 사용할 수 있음  
주어진 데이터셋의 filename속성을 이용해서 카테고리 값을 가져오기 

```python
import pandas as pd
def get_filename_list(newsdata):
    filename_lst = []
    for file in newsdata.filenames:
        filename_temp = file.split("/")[-2:]
        filename = ".".join(filename_temp)
        filename_lst.append(filename)
    return filename_lst

filename_lst = get_filename_list(news_df)
topic_names = ["Topic #" + str(i) for i in range(1,9)]
topic_df = pd.DataFrame(data=doc_topics, columns=topic_names, index=filename_lst)
print(topic_df.head())
```
